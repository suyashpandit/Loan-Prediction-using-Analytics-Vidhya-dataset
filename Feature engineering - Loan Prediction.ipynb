 **Feature engineering:**

1. **One hot encoding**
Categorical features were encoded using one hot encoding and changed to numeric value for model use.

categorical_features = ["Married", "Gender", "Education", "Self_Employed", "Property_Area"]

data = pd.get_dummies(data, columns = categorical_features, drop_first = True)

Encoded categorical features changed to integer values from "yes" and "no" strings.

encoded_cat_features = ["Married_Yes", "Gender_Male", "Education_Not Graduate", "Self_Employed_Yes", "Property_Area_Semiurban", "Property_Area_Urban"]

data[encoded_cat_features] = data[encoded_cat_features].astype(int)


2. **Feature transformation** was done by creating relevant new features from raw features.

Even though some features like Applicant income and Co applicant income stand on their own, combined them into Total income. Creating Estimated Monthly Payment (EMI) based on the domain information that people will less EMI are more likely to not be loan defaulters and so may get approved.

data["Total_income"] = data["ApplicantIncome"] + data["CoapplicantIncome"]

data["EMI"] = data ["LoanAmount"] / data["Loan_Amount_Term"]

3. **Feature selection** was done based on EDA and filter method, by only selecting a portion of the available features.

data.drop(["Loan_ID", "CoapplicantIncome" , "Dependents"], axis = 1, inplace = True)

Loan_id as it is redundant
Coapplicant income as already made Total Income
Dependents column as it doesn't have much effect and has value like 3+

**The goal is to reduce complexity in building the model and improving model performance.**
